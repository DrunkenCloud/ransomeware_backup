```python
import numpy as np
from sklearn.impute import SimpleImputer

# Data with missing values
data_simple_impute = np.array([[1, 2], [np.nan, 3], [7, 6]])
print("Original data for Simple Imputer:\n", data_simple_impute)

# Initialize SimpleImputer with 'mean' strategy
# You can also use 'median', 'most_frequent', or 'constant'
imputer_simple = SimpleImputer(strategy='mean')

# Fit the imputer to the data and transform it
imputed_data_simple = imputer_simple.fit_transform(data_simple_impute)

print("\nImputed data (Simple Imputer):\n", imputed_data_simple)
```

```python
import numpy as np
from sklearn.impute import KNNImputer

# Data with missing values
data_knn_impute = np.array([[1.0, 5.0], [2.0, np.nan], [np.nan, 6.0], [3, 10]])
print("Original data for KNN Imputer:\n", data_knn_impute)

# Initialize KNNImputer with n_neighbors=2
imputer_knn = KNNImputer(n_neighbors=2)

# Fit and transform the data
imputed_data_knn = imputer_knn.fit_transform(data_knn_impute)

print(f"\nImputed data (KNN Imputer with n_neighbors=2s):\n", imputed_data_knn)
print("\nImputed values are the result of considering the single nearest neighbor for each missing value.")
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Data to be scaled
data_minmax = np.array([[-1, 2], [-0.5, 6], [0, 10], [1, 18]])
print("Original data for MinMaxScaler:\n", data_minmax)

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
scaled_data = scaler.fit_transform(data_minmax)

print("\nScaled data (MinMaxScaler):\n", scaled_data)
```

```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np

# T-shirt sizes with a specific order
tshirt_sizes = ['M', 'L', 'S', 'M', 'S', 'L'] # Added more for demonstration
print("Original T-shirt sizes:", tshirt_sizes)

# Define the category order: S < M < L
# The categories need to be provided as a list of lists if there's only one feature.
# Each inner list corresponds to one feature's categories.
category_order = [['S', 'M', 'L']]

# Initialize OrdinalEncoder with the specified categories
encoder = OrdinalEncoder(categories=category_order)

# Fit and transform the data
# Reshape for single feature encoding if it's a 1D array
encoded_tshirt_sizes = encoder.fit_transform(np.array(tshirt_sizes).reshape(-1, 1))

print("\nEncoded T-shirt sizes (S=0, M=1, L=2):\n", encoded_tshirt_sizes.flatten())
```

```python
import pandas as pd

# URL for the Iris dataset
iris_url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

# Load the dataset into a pandas DataFrame
try:
    iris_df = pd.read_csv(iris_url)
    print("Iris dataset loaded successfully!")
except Exception as e:
    print(f"Error loading Iris dataset: {e}")
    iris_df = pd.DataFrame() # Create an empty DataFrame to prevent errors in subsequent cells

# Print the first 5 rows
print("\nFirst 5 rows of the Iris dataset:")
print(iris_df.head())

# Print the last 5 rows
print("\nLast 5 rows of the Iris dataset:")
print(iris_df.tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd # Ensure pandas is imported if running this cell independently

if 'iris_df' in locals() and not iris_df.empty:
    # Set style for plots
    sns.set_style("whitegrid")

    # Create a figure and a set of subplots for multiple scatter plots
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('Scatter Plots of Iris Dataset Features by Species', fontsize=16)

    # Scatter Plot 1: Sepal Length vs. Sepal Width, colored by species
    sns.scatterplot(
        x='sepal_length',
        y='sepal_width',
        hue='species', # Differentiate points by species
        data=iris_df,
        s=100, # Size of the points
        alpha=0.8, # Transparency
        ax=axes[0]
    )
    axes[0].set_title('Sepal Length vs. Sepal Width')
    axes[0].set_xlabel('Sepal Length (cm)')
    axes[0].set_ylabel('Sepal Width (cm)')
    axes[0].legend(title='Species') # Add legend title

    # Scatter Plot 2: Petal Length vs. Petal Width, colored by species
    sns.scatterplot(
        x='petal_length',
        y='petal_width',
        hue='species',
        data=iris_df,
        s=100,
        alpha=0.8,
        ax=axes[1]
    )
    axes[1].set_title('Petal Length vs. Petal Width')
    axes[1].set_xlabel('Petal Length (cm)')
    axes[1].set_ylabel('Petal Width (cm)')
    axes[1].legend(title='Species')

    # Scatter Plot 3: Petal Length vs. Sepal Length, colored by species
    sns.scatterplot(
        x='sepal_length',
        y='petal_length',
        hue='species',
        data=iris_df,
        s=100,
        alpha=0.8,
        ax=axes[2]
    )
    axes[2].set_title('Sepal Length vs. Petal Length')
    axes[2].set_xlabel('Sepal Length (cm)')
    axes[2].set_ylabel('Petal Length (cm)')
    axes[2].legend(title='Species')

    # Adjust layout
    plt.tight_layout(rect=[0, 0.03, 1, 0.96])
    plt.show()
else:
    print("Iris DataFrame is not loaded or is empty. Please run Cell 5 first.")
```
